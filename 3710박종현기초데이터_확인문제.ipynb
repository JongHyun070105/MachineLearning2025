```python
##다음 코드를 실행하여 생성되는 데이터프레임 customer, orders를 이용하여
import pandas as pd
import numpy as np
customer = pd.DataFrame({'customer_id' : np.arange(6),
                    'name' : ['James', 'Elly', 'Tom', 'Givert', 'Aiden', 'Brody'],
                    '나이' : [40, 20, 21, 30, 31, 18]})

orders = pd.DataFrame({'customer_id' : [1, 1, 2, 2, 2, 3, 3, 1, 4, 9],
                    'item' : ['마우스', '충전기', '이어폰', '헤드셋', '전자펜', '키보드', '전자펜', '마우스', '키보드', '케이스'],
                    'quantity' : [1, 2, 1, 1, 3, 2, 2, 3, 2, 1]})
```

##5. '데이터프레임 병합하기'에서 customer, orders 데이터프레임을 사용할 경우 가장 많이 팔린 아이템은 무엇인가요?(merge, groupby, sort_values 활용)
##<<출력형태>>
image.png

```python
### 이곳에 5번 답안 코드를 쓰시오.
merged_df = pd.merge(customer, orders, on='customer_id')

item_sales = merged_df.groupby('item')[['customer_id','name','나이','quantity']].sum()

most_sold_item = item_sales.sort_values(by='quantity', ascending=False)

print(most_sold_item)
```

##6. '데이터프레임 병합하기'에서 customer, orders 데이터 프레임을 사용할 경우 Elly가 가장 구매한 물건은 무엇인가요 ?(merge, groupby, loc 활용)
###<<출력형태>>
###image.png


```python
### 이곳에 6번 답안 코드를 쓰시오.

merged_df = pd.merge(customer, orders, on='customer_id')

elly_data = merged_df.loc[merged_df['name'] == 'Elly']

elly_item_sales = elly_data.groupby('item')[['customer_id','나이','quantity']].sum()

most_purchased_item = elly_item_sales.sort_values(by='quantity', ascending=False)

print(most_purchased_item)
```
