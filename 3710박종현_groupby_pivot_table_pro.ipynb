```python
import pandas as pd

data = {
    '지점': ['강남점', '홍대점', '신촌점', '명동점', '잠실점', '강남점', '홍대점', '신촌점', '명동점', '잠실점',
           '강남점', '홍대점', '신촌점', '명동점', '잠실점', '강남점', '홍대점', '신촌점', '명동점', '잠실점'],
    '월': [1, 1, 2, 2, 3, 1, 2, 3, 1, 2, 3, 1, 2, 3, 1, 2, 3, 1, 2, 3],
    '요일': ['월', '화', '수', '목', '금', '토', '일', '월', '화', '수',
           '목', '금', '토', '일', '월', '화', '수', '목', '금', '토'],
    '음료수': ['아메리카노', '카페라떼', '콜드브루', '바닐라라떼', '초코라떼',
             '아메리카노', '카페라떼', '콜드브루', '바닐라라떼', '초코라떼',
             '아메리카노', '카페라떼', '콜드브루', '바닐라라떼', '초코라떼',
             '아메리카노', '카페라떼', '콜드브루', '바닐라라떼', '초코라떼'],
    '판매수량': [20, 35, 30, 22, 18, 27, 40, 33, 19, 25,
             28, 36, 29, 21, 17, 30, 38, 31, 20, 16],
    '총매출': [80000, 157500, 150000, 114400, 86400,
            108000, 180000, 165000, 98800, 120000,
            112000, 162000, 145000, 109200, 81600,
            120000, 171000, 155000, 104000, 76800]
}

df = pd.DataFrame(data)
df
```

```python
#Q1. 지점별 총매출의 합계를 구하시오.

df.groupby('지점')['총매출'].sum()
```

```python
# Q2. 월별 평균 판매수량을 구하시오.


df.groupby('월')['판매수량'].mean()
```

```python
# Q3. 요일별 총매출 평균을 구하시오.


df.groupby('요일')['총매출'].mean()
```

```python
# Q4. 음료수 종류별로 총 판매수량을 구하시오.

df.groupby('음료수')['판매수량'].sum()
```

```python
# Q5. 각 지점에서 가장 많이 판매된 음료수의 총매출 평균을 구하시오.
# (지점별로 각 음료의 평균 매출을 보고자 함)


df.groupby(['지점','음료수'])['총매출'].mean()
```

```python
# Q6. 지점별·월별 총매출 합계를 구하시오.

df.groupby(['지점','월'])['총매출'].sum()
```

```python
# Q7. 음료수별로 판매된 평균 수량과 평균 총매출을 구하시오.


df.groupby('음료수')[['판매수량','총매출']].mean()
```

```python
# 요일별 지점별 음료수 판매수량 합계를 구하시오.

pd.pivot_table(df,index = ['요일','지점'], columns = '음료수', values = '판매수량', aggfunc='sum', fill_value=0)
```

```python
# 지점별로 월마다의 총매출 합계를 피벗 테이블로 구하시오.

df.pivot_table(index='지점', columns = '월', values='총매출', aggfunc='sum')
```

```python
# 음료수별로 요일마다 평균 판매수량을 구하는 피벗 테이블을 만드시오.

df.pivot_table(index='음료수', columns = '요일', values='판매수량', aggfunc='mean', fill_value=0)
```

```python
# 지점과 월을 기준으로, 평균 총매출과 평균 판매수량을 함께 표시하는 피벗 테이블을 만드시오.


df.pivot_table(index='지점', columns = '월', values=['총매출','판매수량'], aggfunc='mean')
```

```python
# 각 월별로 가장 많이 팔린 음료수의 총 판매수량을 비교할 수 있도록 피벗 테이블을 만드시오.
# (→ 월: index, 음료수: columns)

df.pivot_table(index='월', columns = '음료수', values='판매수량', aggfunc='sum')
```

```python
# 요일별, 지점별 총매출 합계를 피벗 테이블로 구하되, 총합을 포함하여 표시하시오.


df.pivot_table(index = '요일', columns = '지점', values = '총매출' , aggfunc='sum', fill_value=0, margins=True, margins_name = '총합')
```
