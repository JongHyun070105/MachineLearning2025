# SECTION 02. 필요 데이터 변경하기

## 1. 데이터 추가하기

```python
# 기존 데이터프레임의 컬럼을 이용하여 새로운 칼럼 추가하시
# price를 2배로 하는 새로운 컬럼(price2) 만들기

df['price2'] = df['price'] * 2
df['price2']
```

```python
# 기존 컬럼의 연산을 통해 새로운 생성하기
# price1과 price2를 더해서 price3컬럼을 생성

df['price3'] = df['price'] + df['price2']
df['price3']
```

```python
# 새로운 칼럼 원하는 위치에 넣기
# df.insert(loc, column, value, allow_duplicates=False)
# duration 컬럼 값을 10배로 하여 새로운 duration2 컬럼을 만들고, 10번째 칼럼으로 insert하기


df.insert(10, 'duration2', df['duration'] * 10)
df['duration2']
```

## 2. 데이터 삭제하기

```python
# drop 메소드 사용하여 데이터 삭제 하기
# drop 메소드를 사용하여 price3컬럼 삭제하기

df.drop('price3', axis = 1,inplace=True)
df
```

```python
# drop 메소드를 이용하여 첫번째 행을 삭제하기 /axis 속성 이해)

df.drop(axis=0,index=0,inplace=True)
```

```python
# 데이터 삭제 후 새로운 데이터프레임에 저장하기1
# drop메소드를 활용하여 데이터를 지운 후 다른 데이터프레임에 저장하는 방법

df.drop('price2',axis=1,inplace = True)
df
```

```python
# csv파일 불러와서 flight DataFrame (flight)만들기

flight = pd.DataFrame(data = pd.read_csv('./Clean_Dataset.csv'))
```

## 1. 그룹화 하기

### 1) groupby 활용하기

```python
# airline 칼럼 기준으로 그룹화하고, airline_group에 저장하기

airline_group = flight.groupby('airline')
airline_group
```

```python
## 수행결과 데이터 프레임으로 그룹까지 생성됨.
## 그룹화된 결과를 확인하기 위해서는 groups속성을 이용
# 수행 결과 확인하기

airline_group.groups
```

```python
# 데이터 수 확인하기
# count: 데이터개수
# size:집단별 크기
# sum : 데이터 합
# mean, std, var : 평균, 표준편차, 분산
# min, max : 최소값, 최대값

print(airline_group.count())
```

```python
# 최솟값 확인하기
print(airline_group.min())
```

```python
# 평균값 확인하기
num_col = flight.select_dtypes(include=[np.number])
print(num_col.mean())
```

```python
# 특정 칼럼 값만 확인하기
print(num_col['price'].mean())
```

```python
# groupby를 활용하여 다중 인덱싱(multi-indexing)설정하기
#'airline'별 'arrival_time'에 따른 평균값 확인하기

numeric_columns = flight.select_dtypes(include=[np.number])

# 다중 인덱스 설정 후 평균 계산
flight.set_index(['airline', 'arrival_time'])[numeric_columns.columns].groupby(level=[0]).mean()
```

```python
# 여러 개의 칼럼('airline', 'arrival_time')을 groupby하여 새로운 데이터프레임(mul_airline_group) 생성하기

mul_airline_group = flight.groupby(['airline','arrival_time'])

# 새롭게 생성된 mul_airline_group의 평균값 확인하기
mul_airline_group.mean()
```
