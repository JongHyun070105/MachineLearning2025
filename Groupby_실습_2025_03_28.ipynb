```python
import pandas as pd
import numpy as np
```

```python
# flight DataFrame 만들기
flight = pd.read_csv('./Clean_Dataset.csv',encoding='cp949')
flight
```

```python
# airline 기준 그룹 생성
airline_group = flight.groupby('airline')
airline_group.groups
```

```python
airline_group.count()
```

```python
airline_group.mean(numeric_only=True)
```

```python
flight_num = flight.select_dtypes(include='number')
flight_num.groupby(flight['airline']).mean()
```

```python
flight.groupby(['airline','arrival_time']).mean(numeric_only=True)
```

```python
b = flight.select_dtypes(include='number')
b.groupby([flight['airline'],flight['arrival_time']]).mean()
```

```python
# 그룹화를 한 후엔 원하는 데이터만 가져오기
 # AirAsia 항공사의 Evening 시간대에 도착 데이터의 평균값만 필터링

flight.groupby(['airline','arrival_time']).mean(numeric_only=True).loc[[('AirAsia','Evening')]]
```

```python
grouped = flight.groupby(['airline','arrival_time']).mean(numeric_only=True)
grouped.loc[[('AirAsia','Evening')]]
```
