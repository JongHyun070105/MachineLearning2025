```python
import pandas as pd
data = {
    '이름' : ['차은우', '배수지', '이세영', '이도현', '김유정', '서강준', '한소희', '남주혁'],
    '학교' : ['세명고', '한국고', '세명고', '한국고', '은진고', '은진고', '한국고', '세명고'],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90],
    '과학' : [95, 55, 80, 75, 35, 85, 40, 95],
    '사회' : [85, 25, 75, 80, 10, 80, 35, 95],
    '성별' : ['남', '여','여', '남','여','남','여','남'],
    '주소' : ['은평구','','서대문구','','서대문구','강서구','은평구','고양시']
}
data
```
