## Series
- 1차원 데이터 (정수, 실수, 문자열 등등)

```python
import pandas as pd
```

```python
a = pd.Series([10,20,30,40]) # series 생성 - 인덱스와 그에 해당하는 값이 있다.
print(a)
```

```python
print(a[0]) # a의 0번째 인덱스 값 추출
```

```python
# series 생성할 때 인덱스 지정해서 생성
b = pd.Series([10,20,30,40], index=['1번','2번','3번','4번'])
print(b)
```

```python
print(b["2번"])
```

```python
lst = list(range(10,101,10))
print(lst,'\n')
c = pd.Series(lst)
print(c)
```
