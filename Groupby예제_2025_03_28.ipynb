```python
import numpy as np
import pandas as pd

data = {
    '이름' : ['차은우', '배수지', '이세영', '이도현', '김유정', '서강준', '한소희', '남주혁'],
    '학교' : ['세명고', '한국고', '세명고', '한국고', '은진고', '은진고', '한국고', '세명고'],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90],
    '과학' : [95, 55, 80, 75, 35, 85, 40, 95],
    '사회' : [85, 25, 75, 80, 10, 80, 35, 95],
    '성별' : ['남', '여','여', '남','여','남','여','남'],
    '주소' : ['은평구','은평구','서대문구','강서구','서대문구','강서구','','고양시']
}

df = pd.DataFrame(data,index=['1번','2번','3번','4번','5번','6번','7번','8번'])
df.index.name = '학번'
df
df.to_excel('score.xlsx')
```

## 그룹화
##### 동일한 값을 가진 것들끼리 합쳐서 통계치를 계산하기 위해서 사용하는 기능 (엑셀 부분합과 동일)

```python
df.groupby('학교')
```

```python
df.groupby('학교').get_group('세명고')
```

```python
df.groupby('주소').get_group('은평구')
```

```python
df.groupby('학교').mean(numeric_only=True) # 계산 가능한 항목의 평균
```

```python
df.groupby('학교').size() # 각 그룹의 크기
```

```python
# 주소별 각 과목의 평균값 확인
df.groupby('주소').mean(numeric_only=True)
```

```python
# 학교로 그룹화한 후에 세명고에 해당하는 데이터 수
df.groupby('학교').size()['세명고']
```

```python
# 학교별 그룹화를한 뒤에 국어성적의 평균 데이터를 확인
df.groupby('학교').mean(numeric_only=True)['국어']
```

```python
df.groupby('성별')[['국어','영어','수학']].mean()
```

```python
df['학년'] = [3,3,2,1,1,3,2,2]
df
```

```python
df['학과'] = ['인공','보안','보안','인공','인공','인공','보안','게임']
df
```

```python
df.groupby(['학교','학년']).mean(numeric_only=True)
```

```python
# 학교별로 그룹화한 후 학년별 국어, 영어 평균
df.groupby(['학교','학년'])[['국어','영어']].mean()
```

```python
# 주소, 학년별 계산 가능한 항목의 총합 sum()
df.groupby(['주소','학년']).sum()
```

```python
df.groupby('학교')['이름'].count()
```

```python
s = df.groupby('학교')
```

```python
s['학년'].value_counts() # 학교로 그룹화를 한 뒤에 학년별 학생 수를 가져옴
```
